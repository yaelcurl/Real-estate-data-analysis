## Preprocessing

### Handling missing values

#### General Statistics

```python
#checking correlation matrix in order to decide which parameter to use to recover missing values in another
data.corr()
```

```python
#evaluation of mean,median and std in order to choose the right method for missing values recovery 
data_stat=pd.concat([data.median(), data.mean(),data.std(),data.min(),data.max()], axis=1)
data_stat.columns=['median','mean','std','min','max']
data_stat
```

#### ceiling_height

Although std is twice lower than mean, there's no reliable way to recover the missing values (40% of total data) because this column isn't correlated with any other column.

Reason for missing values: probably people were to lazy to measure the ceiling. 
Type: MCAR

#### floors_total

```python
#recovering missing floors_total

#using floor column that has highest correlation of 0.67 with floors_total
#evaluating std to determine whether to use mean or median
data.pivot_table(index='floor',values='floors_total',aggfunc=['median','mean','std']) 
#std is pretty large for floor<6 , it's makes more sense to use median
```

```python
#checking if missing values counts follow some pattern regarding floor
data[data['floors_total'].isna()]['floor'].value_counts()
#seems random
```

```python
#create a dictionary of floors_total median grouped by floor 
floors_total_recovery=data.pivot_table(index='floor',values='floors_total',aggfunc='median')['floors_total'].to_dict()
print(floors_total_recovery)
#map floors_total missing values accordind to dict
data.loc[data['floors_total'].isna(),'floors_total']=data.loc[data['floors_total'].isna(),'floor'].map(floors_total_recovery)
#making sure no floors_total is lower than floor
data.loc[data['floors_total']<data['floor'],'floor'].value_counts()
```

#### living_area

```python
#recovering missing living_area

#using total_area column that has highest correlation of 0.93 with living_area
#evaluating std to determine whether to use mean or median
data.pivot_table(index='total_area',values='living_area',aggfunc=['median','mean','std'])[0:20]
#similar mean and median
```

```python
#checking if missing data is related to one-room and small apartments

#missing values per rooms divided by total entries of rooms
print(data.loc[data['living_area'].isna(),'rooms'].value_counts().sort_index() / data['rooms'].value_counts().sort_index())
#missing values per rooms 
data.loc[data['living_area'].isna(),'rooms'].value_counts()
#data.loc[(data['living_area'].notna()) & (data['rooms']==1),['total_area','living_area']]
```

```python
#create a dictionary of living_area median grouped by total_area 
living_area_recovery=data.pivot_table(index='total_area',values='living_area',aggfunc='median')['living_area'].to_dict()
keys=data.loc[data['living_area'].isna(),'total_area'].round(0)
for key in keys:
    index=bisect.bisect_left(list(living_area_recovery.keys()), key)
    living_area_recovery.setdefault(key, list(living_area_recovery.values())[min(index,len(living_area_recovery.values())-1)])
```

```python
#map living_area missing values accordind to dict
data.loc[data['living_area'].isna(),'living_area']=data.loc[data['living_area'].isna(),'total_area'].round(0).map(living_area_recovery)
data.info()
```

living_area is highly correlated with total_area (0.94 ==> they contain almost same information) and therefore reliably recovered by median values grouped by total_area.

Reason for missing values: as seen above, when grouping by rooms, missing data occurs for about 7% - 12% of total rooms entries count and only for rooms < 9. 
Although most missing data is for small values of rooms (rooms=1 in top), the ratio is pretty much similar for all.
Thus it's probably random and doesn't depend on place size.
The reason might be again people were too lazy to mesure. 

type: MCAR

#### kitchen_area

```python
#checking if missing data is related to one-room and small apartments

#missing values per rooms divided by total entries of rooms
(data.loc[data['kitchen_area'].isna(),'rooms'].value_counts().sort_index() / data['rooms'].value_counts().sort_index())
```

```python
#recovering missing kitchen_area

#create a dictionary of kitchen_area median grouped by total_area 
kitchen_area_recovery=data.pivot_table(index='total_area',values='kitchen_area',aggfunc='median')['kitchen_area'].to_dict()
keys=data.loc[data['kitchen_area'].isna(),'total_area'].round(0)
for key in keys:
    #find the nearest key in the dict for each total_area value
    index=bisect.bisect_left(list(kitchen_area_recovery.keys()), key)
    #set the value in dict according to index 
    kitchen_area_recovery.setdefault(key, list(kitchen_area_recovery.values())[min(index,len(kitchen_area_recovery.values())-1)])
```

```python
#map kitchen_area missing values accordind to dict
data.loc[data['kitchen_area'].isna(),'kitchen_area']=data.loc[data['kitchen_area'].isna(),'total_area'].round(0).map(kitchen_area_recovery)
data.info()
```

kitchen_area is most correlated with total_area (0.6) and therefore recovered by median values grouped by total_area.

Reason for missing values: as seen above, when grouping by rooms, missing data occurs for about 7% - 12% of total rooms entries count for most cases. The reason is probably laziness.
The exceptions are rooms=0 that has 100% missing data for kitchen_area, the reason is probably the fact that the kitchen is part of the living area one-space and there's a difficulty to separate.
rooms=8 has 25%.

## Adding calculations

```python
#adding price per squre meter
data.eval('price_per_m2 = last_price / total_area',inplace=True)
#data['price_per_m2']=(data['last_price'] / data['total_area']).astype('int64')
data['price_per_m2'].astype('int64').describe
```

```python
#adding day of week name , month, year
days_dict={ i : list(calendar.day_name)[i] for i in range(0, 7) }
data['dayofweek_exposition_num']=data['first_day_exposition'].dt.dayofweek
data['dayofweek_exposition']=data['dayofweek_exposition_num'].map(days_dict)
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year
#data.info()
```

```python
#adding floor level
data.loc[data['floor']==data['floors_total'],'floor_level']='last'
data.loc[data['floor']==1,'floor_level']='first'
data['floor_level']=data['floor_level'].fillna('other')
data.loc[:,['floor','floors_total','floor_level']][20:25]
```

```python
#adding area ratios
data.eval('living_from_total_area = living_area / total_area',inplace=True)
data.eval('kitchen_from_total_area = kitchen_area / total_area',inplace=True)
data['living_from_total_area'].round(2)
data['kitchen_from_total_area'].round(2)
```

#### Removing Outliers

```python
data_clean = data
Q={} #a dictionary for lower and upper quartiles and 1.5*IRQ
for col in ['total_area', 'last_price', 'rooms','ceiling_height', 'days_exposition']:
    Q[col]=[data[col].quantile(0.25),\
            data[col].quantile(0.75), \
            1.5*(data[col].quantile(0.75)-data[col].quantile(0.25))]
    #save only rows that satisfy (Q1-1.5IRQ) < col < (Q3+1.5IRQ)
    data_clean = data_clean[(data[col]>(Q[col][0]-Q[col][2])) & (data[col]<(Q[col][1]+Q[col][2]))]
Q
```

```python
#plotting histogram for days exposition after rare values removal
days_exposition_main = data_clean.dropna()['days_exposition']
plt.hist(days_exposition_main,range=(0,days_exposition_main.max()), bins=34)
plt.xlabel('time taken to sell [days]')
plt.ylabel('counts for bins=34')
plt.title('Days Exposition Histogram')
#adding quartiles
days_quartiles = pd.qcut(days_exposition_main,4,retbins=True)
for quartile in days_quartiles[1][0:3]:
    plt.axvline(quartile, color='g', linewidth=1)
#adding mean and median
plt.axvline(days_exposition_main.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(days_exposition_main.median(), color='r', linestyle='dashed', linewidth=1)
plt.figure(figsize=(50,5))
print(Fore.GREEN + 'Days Exposition Quartiles:', list(days_quartiles[1]))
```

After removing 3 < days > 476 and aggregating to 34 bins (~2 weeks / bin) we can see that the number of sales is decreasing with months in general but not continuously. For days values not too large it has peaks in sales at the beginning of each month.
